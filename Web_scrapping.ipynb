1-Write a python program to display all the header tags from wikipedia.org.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
def header_tag(url):
#raise request
    page=requests.get(url)
    page
    soup=BeautifulSoup(page.content)

#fetching header tag(this page have only h1 tag)    
    header_list=[]
    tags=['h1']

    for i in soup.find_all(tags):

        header_list.append(i.text.replace("\n"," ").strip())
    return header_list        
```

```python
#fetching header tags
header_tag('https://www.wikipedia.org/')
```

### 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
and make data frame

```python
def top_100_movie(url):


    movies=requests.get(url)
    movies

    soup1=BeautifulSoup(movies.content)

    name=[]
    for i in soup1.find_all('td',class_='titleColumn'):
        name.append(i.text.replace('\n'," ").strip())

    rating=[]
    for i in soup1.find_all('td',class_='ratingColumn imdbRating'):
        rating.append(i.text.replace('\n'," ").strip())

    year_of_release=[]
    for i in soup1.find_all('span',class_='secondaryInfo'):
        year_of_release.append(i.text)



    movie_data=pd.DataFrame({'name':name,
                     'rating':rating,
                     'year_of_release':year_of_release

    }
    )


    top_100_movie=movie_data.head(100)

    return top_100_movie



    
```

```python
top_100_movie('https://www.imdb.com/chart/top')
```

### 4) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:

a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
def top_team(url):
    top_team=requests.get(url)
    soup=BeautifulSoup(top_team.content)

    #create empty list
    team=[]
    matches=[]
    points=[]
    rating=[]

    #fectching date for all column one by one
    for i in soup.find_all('span',class_='u-hide-phablet',limit=10):
         team.append(i.text.replace('\n',""))

    top=soup.find('td',class_='rankings-block__banner--matches')
    matches.append(top.text)
    for i in soup.find_all('td',class_='table-body__cell u-center-text'):
             matches.append(i.text)
    matches.insert(1,np.nan)
    matches=matches[0:20:2]


    top=soup.find('td',class_='rankings-block__banner--points')
    points.append(top.text)
    for i in soup.find_all('td',class_='table-body__cell u-center-text'):
             points.append(i.text)            

    points=points[0:20:2]

    top=soup.find('td',class_='rankings-block__banner--rating u-text-right')
    rating.append(top.text.replace("\n","").strip())
    for i in soup.find_all('td',class_='table-body__cell u-text-right rating',limit=9):
             rating.append(i.text) 

    rating

    print(len(team),len(matches),len(points),len(rating))

    #create dataframe

    top_team=pd.DataFrame({'team':team,
                           'matches':matches,
                           'points':points,
                           'rating':rating

            }
            )

    return top_team
```

```python
# Team data using url

top_team('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

```python
# function Players(batsman) & their respective team along with rating

def top_player(url):
        top_player=requests.get(url)
        soup=BeautifulSoup(top_player.content)

        player=[]

    #top 1 player fetch seperately because of different class
        top=soup.find('div',class_='rankings-block__banner--name-large')
        player.append(top.text)
        for i in soup.find_all('td',class_='table-body__cell rankings-table__name name'):
                player.append(i.text.replace('\n',""))


        team=[]

        top=soup.find('div',class_='rankings-block__banner--nationality')
        team.append(top.text.replace("\n",""))
        for i in soup.find_all('span',class_='table-body__logo-text'):
            team.append(i.text)


        ratings=[]

        top=soup.find('div',class_='rankings-block__banner--rating')
        ratings.append(top.text.replace("\n",""))
        for i in soup.find_all('td',class_='table-body__cell rating'):
            ratings.append(i.text)    


        player_data=pd.DataFrame({'player':player,
                             'team':team,
                             'ratings':ratings

        }
        )

        top_player=player_data.head(10)
        return top_player
```

## 6) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content
and the code for the video from the link for the youtube video from the post.

```python
#send request
def CoreyMS(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)

    #scrap headings
    heading=[]

    tags=['h2']

    for i in soup.find_all(tags):
        heading.append(i.text)


    #scrap date

    date=[]

    for i in soup.find_all('time',class_='entry-time'):
        date.append(i.text)

    #scrap content
    content=[]

    for i in soup.find_all('div',class_='entry-content'):
        content.append(i.text.replace('\n',""))

    #scrap vid_link
    vid_link=[]

    vid_link=[]
    for i in soup.find_all('iframe'):
        vid_link.append(i['src'])
    #no video at index 4 so adding nan
    vid_link.insert(4,np.nan)    

    CoreyMS=pd.DataFrame({'heading':heading,
                          'date':date,
                          'content':content,
                          'vid_link':vid_link

    }
    )

    return CoreyMS
```

```python
#data fetching
CoreyMS('https://coreyms.com/')
```

## 7) Write a python program to scrape house details from mentioned URL. It should include house title, location,
area, EMI and price from nobroker.in.

```python
#send request
def house_data(hurl):
    page=requests.get(hurl)
    soup=BeautifulSoup(page.content)

#scrap house title
    house_title=[]

    for i in soup.find_all('h2'):
        house_title.append(i.text)
    
    
#scrap location

    location=[]

    for i in soup.find_all('div',class_='nb__1EwQz'):
         location.append(i.text)


#scrap area

    area=[]

    for i in soup.find_all('div',class_='nb__FfHqA'):
         area.append(i.text)


#scrap emi
    emi=[]

    for i in soup.find_all('div',class_='font-semi-bold heading-6',id='roomType'):
         emi.append(i.text)

#scrap price
    price=[]

    for i in soup.find_all('div',class_='font-semi-bold heading-6'):
             price.append(i.text)

    price=price[2::3]

#creating dataframe
    house_data=pd.DataFrame({'house_title':house_title,
                          'location':location,
                           'price':price,  
                          'area':area,
                          'emi':emi

    }
    )

    return house_data
```

## 10) Write a python program to scrape monument name, monument description, image URL about top 10 monuments
from puredestinations.co.uk.
